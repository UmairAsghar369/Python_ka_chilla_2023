```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
kashti = sns.load_dataset("titanic")
kashti
```

```python
kashti.info()
```

```python
# Null values dekne ke lie command
kashti.isnull().sum()
```

### yahan pr 3 columns me null values, to hum inko data wrangling me ahtiyat se deal krein ge 

```python
# plot
kashti.survived.value_counts().plot(kind="bar")
```

```python
# bar plot
sns.countplot(x="survived",hue='sex',data=kashti)
```

```python
# scatter plot
sns.scatterplot(x="age",y="fare",hue="pclass",data=kashti)
```

```python
# bar plot
sns.barplot(x="sex",y="fare",hue="pclass",data=kashti)
```

```python
sns.pairplot(kashti)
# har aik variable ko dusray variable k sath compare krta hai
```

```python
# histplot
kashti.hist(figsize=(10,10))
# jitne numeric variables hain unka histogram bna de ga
```

```python
# box plot
sns.boxplot(x="pclass",y="age",data=kashti)
```

```python
sns.boxenplot(x="pclass",y="age",data=kashti)
```

```python

g = sns.FacetGrid(kashti, hue='survived', col='pclass',margin_titles=True,
                  palette={0:'red',1:'green'})
g= g.map(plt.scatter, 'fare', 'age',edgecolor='w').add_legend()
```

```python
g = sns.FacetGrid(kashti, hue='survived', col='sex',margin_titles=True,
                  palette={0:'red',1:'green'})
g= g.map(plt.scatter, 'fare', 'age',edgecolor='w')
```

```python
sns.stripplot(x="sex",y="fare",hue='class', data=kashti)
```

```python
# heat map
corr = kashti.select_dtypes(include='number').corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
sns.violinplot(x="sex",y="fare",hue='class', data=kashti) 
```
