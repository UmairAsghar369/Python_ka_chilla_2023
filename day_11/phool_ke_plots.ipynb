```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
phool = sns.load_dataset("iris")
phool.head()
```

```python
phool.info()
```

```python
phool.isnull().sum()
```

```python
sns.countplot(x="species",hue='sepal_length',data=phool)
```

```python
sns.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=phool)
```

```python
sns.scatterplot(x="petal_length",y="petal_width",hue="species",data=phool)
```

```python
sns.barplot(x="petal_length",y="petal_length",hue="species",data=phool)
```

```python
sns.pairplot(phool)
```

```python
phool.hist(figsize=(10,10))
```

```python
sns.boxplot(x="petal_width",y="sepal_length",data=phool)
```

```python
sns.boxenplot(x="petal_width",y="sepal_length",data=phool)
```
