# 03- How to rename columns

```python
df = pd.DataFrame({'A col':[1,2,3,7,8,9],'B col':[4,5,6,10,11,12]})
df
```

```python
df.rename(columns={'A col':'Col_A', 'B col':'Col_B'}, inplace=True)
df
```

```python
#Another method to rename columns
df.columns = ['col_AA', 'col_BB']
df
```

```python
# To replace any character, string
df.columns = df.columns.str.replace('_','-')
df
```

```python
# Adding prefix and suffix to column names
df = df.add_prefix('pre_').add_suffix('_suf')
df
```

```python
df.columns = ['A', 'B']
df
```
