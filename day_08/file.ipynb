## 16 - Splitting a string into multiple columns

```python
import pandas as pd
df = pd.DataFrame({'name':['Umair Asghar', 'Ali Khan','Sajjad Haider','Umar Farooq'],
                   'location':['Lahore, pakistan', 'Sargodha, pakistan', 'Islamabad, pakistan', 'Chicago, USA']})
df
```

```python
# adding those splits into new columns
df[['First_name','last_name']] = df.name.str.split(' ',expand = True)  # split by space
df
```

```python
df[['city','country']] = df.location.str.split(',', expand=True)
df
```

```python
# Refine data manipulation
# ab apni marzi se columns select kr skte hain 

df = df[['First_name','last_name','city','country']]
df
```
