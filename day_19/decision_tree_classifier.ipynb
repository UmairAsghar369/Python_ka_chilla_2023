# Decision Tree Classifier

1. import the necessory liberaries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns       
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
df =  sns.load_dataset("titanic")
```

```python
X = df[["pclass", "sex", "age", "sibsp", "parch", "fare"]]
y = df["survived"]
```

```python
X = pd.get_dummies(X, columns=["sex"])  # Convert categorical variable to dummy variables
# 
```

```python
X.isnull().sum()
```

```python
X.age.fillna(value = X['age'].mean(), inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_prep = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_prep))
```

```python
tree.plot_tree(model.fit(X,y), filled = True)
plt.show()
plt.savefig("tree.png")
```
