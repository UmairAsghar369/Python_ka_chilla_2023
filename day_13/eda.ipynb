 # EDA
  Exploratory data analysis (EDA) is a method used to 'analyze' and 'summarize' datasets. In python, you can use the pandas library to perform EDA on a dataset.


```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# import data set
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
# summarize (summary statistics)
df.describe()
```

```python
# find null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
# finding unique values  (number of unique values)
df.nunique()
```

```python
df.columns
```

```python
# Visualize the nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```
