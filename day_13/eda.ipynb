 # EDA
  Exploratory data analysis (EDA) is a method used to 'analyze' and 'summarize' datasets. In python, you can use the pandas library to perform EDA on a dataset.


```python
# import liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# import data set
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
# summarize (summary statistics)
df.describe()
```

```python
# find null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean() * 100
```

```python
# finding unique values  (number of unique values)
df.nunique()
```

```python
df.columns
```

```python
# Visualize the nan values in heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df.drop(columns='deck', inplace=True)
```

```python
df.shape
```

```python
sns.countplot(data=df, x='survived' )
```

```python
sns.barplot(data=df, x='survived', y='age', hue='who')
```

barplot ke andr x axies pe categorical or y axis pe numeric variable 

```python
sns.countplot(x='pclass', data=df)
```

```python
# data distribution
sns.histplot(data=df, x='age', bins=30)
 # bins jitni ziada data distribution smjne me utni asani
```

data distribution : at a specific data point(people) how many data units are there.

```python
df['age'].hist(bins=30, alpha=0.7)
```

isme hum dek skte hain ke man me outliers hain

```python
#  scatterplot
sns.scatterplot(x='age', y='fare', data=df)
```

jab dono jaga pe numeric variable le aien ge to scatterplot bane ga 

# Hum ye graph bna rai he data ko visulize krne ke lie, 
## isme hum clearly dek sakte hain ke outliers hain ke nhi take phir un outliers ko remove kr skein

```python
sns.scatterplot(x='age', y='fare', data=df, hue='pclass')
plt.ylim(0,300) 
```

isme hum ne ylim lga ke ke outliers ko remove kr dea grapgh se or hue me catogorical variable add kia 
logically ye samjte hue ke kraia (fair) class oe depend krta (nake gender or deck ki catogery pe)

```python
sns.pairplot(df, hue ='survived')
```

```python
# stacked barchart/plot
df.groupby(['pclass', 'sex'])['survived'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
```

```python
# swarm
sns.swarmplot(data= df, x='age', y='fare')
```

## swarm plot bhi scatter plot ke jaisa hi ha 
## iska fiada ye ke agr barplot ke uper swarmplot bna lein ge to uske data points show krne lge ga.

```python
sns.swarmplot(data = df, x= 'survived', y='fare', hue='who')
```

```python
sns.catplot(x='pclass', y='fare', hue='embarked', kind='point',data=df)
```

```python
sns.implot(data=df, x='age', y='')
```

```python
# grouping
df.columns
```

```python
df['age_groups'] = pd.cut(df['age'],[0,10,18,30,50,80], 
    labels=['Kids','Child','Youth','Middle aged', 'Old'])
```

```python
df.head()
```

```python
df.groupby("age_groups")['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived',vars=['age','fare','pclass'])
```
